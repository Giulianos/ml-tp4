```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('out.csv')

plt.plot(df['avgWordLength'])
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

km = KMeans(
    n_clusters=5, init='random',
    n_init=1, max_iter=1000, 
    tol=1e-100, random_state=5
)
scaler = StandardScaler()
scaler.fit(df)

y_km = km.fit_predict(scaler.transform(df))
plt.plot(y_km)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

from sklearn.neighbors import kneighbors_graph
connectivity = kneighbors_graph(df, n_neighbors=12, include_self=True)
ac = AgglomerativeClustering(n_clusters=5, connectivity=connectivity, linkage='ward')

scaler = StandardScaler()
scaler.fit(df)

y_km = ac.fit_predict(scaler.transform(df))
plt.plot(y_km)
plt.show()
```
