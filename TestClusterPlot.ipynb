# K-Medias

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('out_kmeans.csv')

calderaro = df[0:12]
fonteveccia = df[12:22]
pagni = df[22:32]
vanderkooy = df[32:42]
verbitsky = df[42:52]

bins = [0, 1, 2, 3, 4, 5 ]

plt.figure(figsize=(16,8))

p1 = plt.hist([
    calderaro['predCluster'],
    fonteveccia['predCluster'],
    pagni['predCluster'],
    vanderkooy['predCluster'],
    verbitsky['predCluster']], bins=bins, stacked=True, width=0.5)

plt.legend(('Calderaro', 'Fonteveccia', 'Pagni', 'Van Der Kooy', 'Verbitsky'))

plt.show()
```

# Kohonen

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

net_rows = 25
net_cols = 25

df = pd.read_csv('out_kohonen.csv')

calderaro = df[0:12]
fonteveccia = df[12:22]
pagni = df[22:32]
vanderkooy = df[32:42]
verbitsky = df[42:52]

heat = np.zeros((net_rows,net_cols))

for d in calderaro['predCluster']:
    row = int(d/net_cols)
    col = int(d%net_cols)
    heat.itemset((row,col),heat.item((row,col))+1.0)

fig, axs = plt.subplots(3, 2)
fig.set_size_inches(16,16)

sns.heatmap(heat, ax=axs[0,0])
axs[0,0].set_title('Calderaro')

heat = np.zeros((net_rows,net_cols))

for d in fonteveccia['predCluster']:
    row = int(d/net_cols)
    col = int(d%net_cols)
    heat.itemset((row,col),heat.item((row,col))+1.0)
    
sns.heatmap(heat, ax=axs[0,1])
axs[0,1].set_title('Fonteveccia')


for d in pagni['predCluster']:
    row = int(d/net_cols)
    col = int(d%net_cols)
    heat.itemset((row,col),heat.item((row,col))+1.0)
    
sns.heatmap(heat, ax=axs[1,0])
axs[1,0].set_title('Pagni')

for d in vanderkooy['predCluster']:
    row = int(d/net_cols)
    col = int(d%net_cols)
    heat.itemset((row,col),heat.item((row,col))+1.0)
    
sns.heatmap(heat, ax=axs[1,1])
axs[1,1].set_title('Van der Kooy')

for d in verbitsky['predCluster']:
    row = int(d/net_cols)
    col = int(d%net_cols)
    heat.itemset((row,col),heat.item((row,col))+1.0)
    
sns.heatmap(heat, ax=axs[-1,0])
axs[-1,0].set_title('Verbitsky')

axs[-1,-1].axis('off')

plt.show()
```
