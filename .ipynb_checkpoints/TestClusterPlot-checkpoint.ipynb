```python
import pandas as pd

df = pd.read_csv('out_kmeans.csv')

df
```

```python
import matplotlib.pyplot as plt

calderaro = df[0:12]
fonteveccia = df[12:22]
pagni = df[22:32]
vanderkooy = df[32:42]
verbitsky = df[42:52]

bins = [0, 1, 2, 3, 4 ]

p1 = plt.hist(calderaro['predCluster'], bins=bins, stacked=True, width=0.5)
p2 = plt.hist(fonteveccia['predCluster'], bins=bins, stacked=True, width=0.5)
p3 = plt.hist(pagni['predCluster'], bins=bins, stacked=True, width=0.5)
p4 = plt.hist(vanderkooy['predCluster'], bins=bins, stacked=True, width=0.5)
p5 = plt.hist(verbitsky['predCluster'], bins=bins, stacked=True, width=0.5)

plt.legend(('Calderaro', 'Fonteveccia', 'Pagni', 'Van Der Kooy', 'Verbitsky'))

plt.show()
```
